```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats
```

```python
df2 = pd.read_csv('Data/data_1_var_82.txt', sep=',', header=None).T[0]
```

```python
bins = 50

x2 = np.linspace(df2.min(), df2.max(), num=bins)
counts2, bins2 = np.histogram(df2, bins=bins)
mean2 = df2.mean()
std2 = df2.std()

n = len(counts2)
h = (df2.max() - df2.min()) / bins
ncounts2 = np.empty(n)

for i in range(n):
    u = (x2[i] - mean2) / std2
    phi = 1 / np.sqrt(2 * np.pi) * np.exp(-(u**2)/2)
    ncounts2[i] = sum(counts2) * h / std2 * phi
    
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df2, bins=bins2, label='histogram')
ax.plot(x2, counts2, color='y', marker='o', linewidth=2.0, label='polygon')
ax.plot(x2, ncounts2, color='r', marker='o', linewidth=2.0, label='polygon')
fig.show()
```

```python
observed_crit = 0
for i in range(n):
    observed_crit = observed_crit + (counts2[i] - ncounts2[i]) ** 2 / ncounts2[i]
print(observed_crit)
```

```python
a = 0.05
k = n - 3
crit = stats.chi2.ppf(1 - a, k)
okay = observed_crit < crit
print(observed_crit, '<' if okay else '>' ,crit, 
      'then', 'it is normal' if okay else 'it is not normal')
```
