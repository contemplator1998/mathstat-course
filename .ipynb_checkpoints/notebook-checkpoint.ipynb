```python
#Вычислить выборочное среднее, выборочную дисперсию, исправленную дисперсию, 
# размах выборки, медиану, квартили и выборочную квантиль уровня 1/3. Построить 
# гистограмму и полигон частот. На этом же рисунке построить график плотности 
# нормального закона распределения со средним равным выборочному среднему и 
# дисперсией равной исправленной дисперсии. Построить на одном рисунке графики 
# эмпирической функции распределения и функции распределения нормального закона 
# с теми же параметрами как и выше.
# Вычислить доверительные интервалы для среднего и дисперсии выборке считая, что 
# данные имею нормальное распределение. Доверительную вероятность взять 0.99.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats
```

```python
df = pd.read_csv('Data/data_1_var_81.txt', sep=',', header=None).T[0]
```

```python
mean = sum(df) / len(df)
not_corrected_var = sum((df - mean) ** 2) / (len(df))
var = sum((df - mean) ** 2) / (len(df)-1)
std = var ** 0.5
median = df.median()
q25 = df.quantile(0.25)
q33 = df.quantile(1/3)
q50 = df.quantile(0.5)
q75 = df.quantile(0.75)
q100 = df.quantile(1.0)
```

```python
print(f'выборочное среднее: {mean}')
print(f'выборочная дисперсия: {not_corrected_var}')
print(f'исправленная дисперсия: {var}')
print(f'размах выборки: {df.max() - df.min()}')
print(f'медиана: {median}')
print(f'квартиль 25%: {q25}')
print(f'квантиль уровня 1/3: {q33}')
print(f'квартиль 50%: {q50}')
print(f'квартиль 75%: {q75}')
print(f'квартиль 100%: {q100}')
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(df)
ax.axhline(y=mean, color='red', linestyle='dashed')
ax.axhline(y=median, color='green', linestyle='dashed')
ax.legend(['data', f'sample mean = {mean}', f'sample median = {median}'])
fig.show()
```

```python
bins = 50
counts, values = np.histogram(df, bins=bins)

n = len(counts)
h = (df.max() - df.min()) / bins
ncounts = np.empty(n)

for i in range(n):
    u = (values[i] - mean) / std
    phi = 1 / np.sqrt(2 * np.pi) * np.exp(-(u**2)/2)
    ncounts[i] = sum(counts) * h / std * phi

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df, bins=bins, label='histogram')
ax.plot(values[:-1] + h/2, counts, color='y', marker='o', linewidth=2.0, label='polygon')
ax.plot(values[:-1], ncounts, 'r', label='normal')
ax.grid(True)
ax.legend(loc='right')
ax.set_title('Histogram & Polygon & Normal approximation')
ax.set_xlabel('Input values')
ax.set_ylabel('Number of occurrences')
fig.show()
```

```python
bins = 50
x = np.linspace(df.min(), df.max(), bins)
nd = stats.norm(loc=mean, scale=std)
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, nd.cdf(x), 'k--', linewidth=1.5, label='Theoretical')
ax.hist(df, bins, density=True, histtype='step',
            cumulative=True, label='Empirical')
ax.grid(True)
ax.legend(loc='right')
ax.set_title('Cumulative step histograms')
ax.set_xlabel('Input values')
ax.set_ylabel('Likelihood of occurrence')
fig.show()
```

```python
def mean_ci_t_score_manual(data, confidence=0.99):
    alpha = 1-confidence               # significance level
    df = len(data) - 1                  # degress of freedom
    t = stats.t.ppf(1 - alpha/2, df)   # t-critical value for 99% 
    s = np.std(data, ddof=1)            # sample standard deviation
    n = len(data)

    lower = np.mean(data) - (t * s / np.sqrt(n))
    upper = np.mean(data) + (t * s / np.sqrt(n))
    return (lower, upper)
```

```python
def mean_ci_t_score(data, confidence=0.99):
    return stats.t.interval(confidence, len(data) - 1, loc=np.mean(data), scale=stats.sem(data))
```

```python
def var_ci(data, confidence=0.99):
    alpha = 1 - confidence     # significance level
    n = len(data)               # sample sizes
    s2 = np.var(data, ddof=1)   # sample variance
    df = n - 1                 # degrees of freedom

    upper = (n - 1) * s2 / stats.chi2.ppf(alpha / 2, df)
    lower = (n - 1) * s2 / stats.chi2.ppf(1 - alpha / 2, df)
    return (lower, upper)
```

```python
assert(mean_ci_t_score_manual(df) == mean_ci_t_score(df))
print(f'доверительный интервал для среднего: {mean_ci_t_score(df)}')
print(f'доверительный интервал для дисперсии: {var_ci(df)}')
```

```python
# source: https://www.matburo.ru/Examples/Files/ms_pg_3.pdf
def chi_square_test_norm(data, bins=50):
    x = np.linspace(data.min(), data.max(), num=bins)
    counts, values = np.histogram(data, bins=bins)
    mean = data.mean()
    std = data.std()

    n = len(counts)
    h = (data.max() - data.min()) / bins
    ncounts = np.empty(n)

    for i in range(n):
        u = (x[i] - mean) / std
        phi = 1 / np.sqrt(2 * np.pi) * np.exp(-(u**2)/2)
        ncounts[i] = sum(counts) * h / std * phi

    fig, ax = plt.subplots(figsize=(8, 4))
    ax.hist(data, bins=bins, label='histogram')
    ax.plot(x, counts, color='y', marker='o', linewidth=2.0, label='polygon')
    ax.plot(x, ncounts, color='r', marker='o', linewidth=2.0, label='polygon')
    fig.show()
    
    observed = 0
    for i in range(n):
        observed = observed + (counts[i] - ncounts[i]) ** 2 / ncounts[i]
    a = 0.05
    k = n - 3
    critical = stats.chi2.ppf(1 - a, k)
    return (observed, critical)
```

```python
observed, critical = chi_square_test_norm(df2, 100)
if observed > critical: 
    print(f"Hypothesis rejected with 5% significance. {observed} > {critical}")
```

```python
# source http://www.machinelearning.ru/wiki/index.php?title=Критерий_Лемана-Розенблатта
def rank(data,m):
    total=0
    for i in range(len(data)):
        total+=pow((len(data[data<=data[i]])-i-1),2)
    return (1/m)*total
    
def lehmann_rosenblatt(data,data2):
    n=len(data)
    m=len(data2)
    first = rank(data,m)
    second = rank(data2,n)
    total = first + second + 1/6
    obs = (n*m)/(n+m)*(1/(n*m)*total - 2/3)
    crit = 0.46 # for 0.05 alpha, see source ref
    return (obs, crit)
```

```python
obs, crit = lehmann_rosenblatt(df,df2)
if obs > crit:
    print("Hypothesis rejected with 5% significance.")
```

```python
# source: http://scask.ru/q_book_stat1.php?id=144
def chi_square_test(data, data2, bins=50, a=0.05, data_range = (-20, 45)):
    y, x = np.histogram(data, bins=bins, range=data_range)
    y2, x2 = np.histogram(data2, bins=bins, range=data_range)
    
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.hist(data, bins=bins, label='histogram')
    ax.plot(x2[1:], y2, color='y', marker='o', linewidth=2.0, label='polygon')
    fig.show()
    
    n1 = sum(y)
    n2 = sum(y2)
    s = 0
    for i in range(len(y)):
        if y[i] + y2[i] != 0:
            s = s + ((y[i] / n1 - y2[i] / n2) ** 2 / (y[i] + y2[i]))
    obs = n1 * n2 * s
    k = len(y) - 1
    crit = stats.chi2.ppf(1 - a, k)
    return (obs, crit)
```

```python
# source: https://www.matburo.ru/Examples/Files/ms_pg_22.pdf
def chi_square_test2(data, data2, bins=50, a=0.05, data_range = (-20, 45)):
    y, x = np.histogram(data, bins=bins, range=data_range)
    y2, x2 = np.histogram(data2, bins=bins, range=data_range)
    total = y + y2
    yt = total / sum(total) * len(data)
    y2t = total / sum(total) * len(data2)
    chi = 0
    for i in range(len(y)):
        if yt[i] != 0:
            chi = chi + (y[i] - yt[i]) ** 2 / yt[i]
    for i in range(len(y2)):
        if y2t[i] != 0:
            chi = chi + (y2[i] - y2t[i]) ** 2 / y2t[i]
    k = bins - 1
    critical = stats.chi2.ppf(1 - a, k)
    return (chi, critical)
```

```python
chi, critical = chi_square_test(df, df2)
chi2, critical2 = chi_square_test2(df, df2)
assert chi == chi2 and critical == critical2
if chi > critical:
    print("Hypothesis rejected with 5% significance.")
```
